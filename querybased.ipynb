```python
# üìÇ Step 2: Upload your PDF
from google.colab import files
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]
```

```python
from pdf2image import convert_from_path
images = convert_from_path(pdf_path, first_page=47, last_page=56)
print(f"‚úÖ Converted {len(images)} pages to images.")
```

```python
import pytesseract

full_text = ""
custom_config = r'--oem 3 --psm 6'

for i, img in enumerate(images):
    text = pytesseract.image_to_string(img, config=custom_config)
    full_text += f"\n\n--- PAGE {i+1} ---\n" + text
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
# üß† Step 6: Utility to extract keywords from question
def extract_keywords(question):
    stopwords = {"what", "is", "the", "in", "of", "a", "an", "to", "and", "does", "do", "mean"}
    return [word.lower() for word in question.split() if word.lower() not in stopwords]
```

```python
def ask_question_with_page(question):
    from transformers import pipeline

    qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

    best_answer = ""
    best_score = 0
    best_page = None

    for i, img in enumerate(images):
        page_text = pytesseract.image_to_string(img)
        result = qa_pipeline(question=question, context=page_text)
        if result['score'] > best_score:
            best_score = result['score']
            best_answer = result['answer']
            best_page = i + 1

    print(f"\nüîé Question: {question}")
    if best_page:
        print(f"üìÑ Most Relevant Page: {best_page}")
        print(f"‚úÖ Answer: {best_answer}")
    else:
        print("‚ùå Could not find a relevant answer.")

```

```python

ask_question_with_page("What is the meaning of a == b?")
```

```python

# Example
ask_question_with_page("What is functions?")
```
